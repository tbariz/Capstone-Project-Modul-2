### 2.Data Cleaning
- Membersihkan data dari nilai null
- Menghapus Data Duplikat
- Mengurangi data yang tidak diperlukan/data tidak masuk akal
- Menambahkan kolom yang diperlukan
- Merubah nama kolom agar lebih mudah di proses

```python
# isi nilai null pada kolom income dengan median/nilai tengah
df['Income'] = df['Income'].fillna(df['Income'].median())
df['Income'].isnull().sum()
```

```python
# cek apakah ada data duplikat
df.duplicated().sum()

# jika ada data duplikat bisa menggunakan syntax dibawah ini
# df.drop_duplicates(inplace=True)
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
outlier_report = []

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    outlier_report.append((col, int(n_out)))

pd.DataFrame(outlier_report, columns=['Column', 'Outlier Count']).sort_values('Outlier Count', ascending=False)
```

```python
# menambahkan kolom usia pada dataframe
df['Age'] = 2025 - df['Year_Birth']
df.head()
```

```python
# Merapikan kolom dengan memindahkan kolom 'Age' yang baru dibuat ke sebelah koloh 'Year_Birth'
kolom = df.columns.tolist()
kolom.remove('Age')
kolom.insert(2, 'Age')
df = df[kolom]
df.head()
```

```python
# Merapihkan urutan kolom campaign dari campaign 1 (AcceptedCmp1) sampai campaign 5 (AcceptedCmp5)

campaign_cols = [col for col in df.columns if 'AcceptedCmp' in col]
other_cols = [col for col in df.columns if col not in campaign_cols]

sorted_campaign = sorted(campaign_cols, key=lambda x: int(x.replace('AcceptedCmp', '')))

df = df[other_cols + sorted_campaign]

df.head()
```

```python
# Menghapus kolom yang dirasa kurang perlu
df = df[(df['Age'] > 18) & (df['Age'] < 100)]
df
```

```python
# merubah kolom tanggal
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df.head()
```

```python
# Merubah beberapa nama kolom menjadi lebih jelas dang singkat agar lebih mudah melakukan analisis
df = df.rename(columns={
    'MntWines': 'Wines',
    'MntFruits': 'Fruits',
    'MntMeatProducts': 'Meats',
    'MntFishProducts': 'Fish',
    'MntSweetProducts': 'Sweet',
    'MntGoldProds': 'Gold',
    'NumDealsPurchases': 'DiscPurchase',
    'NumWebPurchases': 'WebPurchase',
    'NumCatalogPurchases': 'CatalogPurchase',
    'NumStorePurchases': 'StorePurchase',
    'NumWebVisitsMonth': 'WebVisitsMonth',
    'AcceptedCmp1': '1Campaign',
    'AcceptedCmp2': '2Campaign',
    'AcceptedCmp3': '3Campaign',
    'AcceptedCmp4': '4Campaign',
    'AcceptedCmp5': '5Campaign',
})

df.head()
```

### 7. Inferential Test
- 7.1 Independent T-test
- 7.2 Anova
- 7.3 Chi Square


```python
from scipy.stats import ttest_ind, f_oneway, chi2_contingency

# 7.1) Independent T-test: bandingkan Total_Spending antara pelanggan yang punya anak vs tidak
if 'Kidhome' in df.columns and 'Teenhome' in df.columns and df['Total_Spending'].notna().sum() > 0:
    # Buat kolom HasChild: 1 kalau punya anak/teen di rumah, 0 kalau tidak
    df['HasChild'] = np.where((df['Kidhome'] + df['Teenhome']) > 0, 1, 0)

    g1 = df.loc[df['HasChild'] == 1, 'Total_Spending'].dropna()
    g0 = df.loc[df['HasChild'] == 0, 'Total_Spending'].dropna()

    if len(g1) > 2 and len(g0) > 2:
        t, p = ttest_ind(g1, g0, equal_var=False)
        print('[T-test] Total Pengeluaran pelanggan yang memiliki anak dirumah dan yang tidak (1 = punya anak, 0 = tidak)')
        print(f'T-statistic: {t:.3f}, p-value: {p:.4f}')
        print('Interpretasi: p<0.05 → perbedaan signifikan rata-rata Total Pengeluaran antara pelanggan dengan anak dan tanpa anak.\n')
    else:
        print('[T-test] Data tidak cukup untuk kedua grup.\n')
else:
    print('[T-test] Kolom Kidhome/Teenhome atau Total Pengeluaran tidak tersedia/valid.\n')
```

```python
# 7.2) ANOVA: bandingkan Total Pengeluaran antar level Education
if 'Total_Spending' in df.columns and 'Education' in df.columns and df['Education'].nunique() >= 3:
    groups = [g['Total_Spending'].dropna() for _, g in df.groupby('Education') if len(g['Total_Spending'].dropna()) > 2]
    if len(groups) >= 3:
        F, p = f_oneway(*groups)
        print('[ANOVA] Total Pengeluaran by Education')
        print(f'F-statistic: {F:.3f}, p-value: {p:.4f}')
        print('Interpretasi: p<0.05 → ada perbedaan signifikan Total Pengeluaran antar level Education.\n')
    else:
        print('[ANOVA] Kelompok Total Pengeluaran per Education kurang dari 3 atau data tidak cukup.\n')
else:
    print('[ANOVA] Kolom Total Pengeluaran/Education tidak tersedia.\n')
```

```python
# 7.3) Chi-square: asosiasi Complain vs Response
if 'Complain' in df.columns and 'Response' in df.columns:
    ct = pd.crosstab(df['Complain'], df['Response'])
    if ct.shape[0] >= 2 and ct.shape[1] >= 2:
        chi2, p, dof, exp = chi2_contingency(ct)
        print('[Chi-square] Complain vs Response')
        print(f'Chi2: {chi2:.3f}, dof: {dof}, p-value: {p:.4f}')
        print('Interpretasi: p<0.05 → ada asosiasi signifikan antara komplain dan respon kampanye.')
    else:
        print('[Chi-square] Tabel kontingensi tidak mencukupi (butuh setidaknya 2x2).')
else:
    print('[Chi-square] Kolom Complain/Response tidak tersedia.')
```
